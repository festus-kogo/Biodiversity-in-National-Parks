```python
# Conservation statuses
species_of_concern = 0
endangered = 0
threatened = 0
in_recovery = 0

endangered_species = [] # list
endangered_species_cnt = 0 # counter

endangered_mammals = 0
endangered_birds = 0
endangered_reptile = 0
endangered_fishes = 0
endangered_amphibians = 0
endangered_vascular_plants = 0
endangered_nonvascular_plants = 0

in_recovery_mammals = 0
in_recovery_birds = 0
in_recovery_reptile = 0
in_recovery_fishes = 0
in_recovery_amphibians = 0
in_recovery_vascular_plants = 0
in_recovery_nonvascular_plants = 0

threatened_mammals = 0
threatened_birds = 0
threatened_reptile = 0
threatened_fishes = 0
threatened_amphibians = 0
threatened_vascular_plants = 0
threatened_nonvascular_plants = 0

for i in range(0, len(df['conservation_status'])):
    if df['conservation_status'].iloc[i] == 'Species of Concern':
        species_of_concern += 1
    elif df['conservation_status'].iloc[i] == 'Endangered':
        endangered += 1
        if df['category'].iloc[i] == 'Mammal':
            endangered_mammals += 1
        elif df['category'].iloc[i] == 'Bird':
            endangered_birds += 1
        elif df['category'].iloc[i] == 'Reptile':
            endangered_reptile += 1
        elif df['category'].iloc[i] == 'Fish':
            endangered_fishes += 1
        elif df['category'].iloc[i] == 'Vascular Plant':
            endangered_vascular_plants += 1
        elif df['category'].iloc[i] == 'Nonvascular Plant':
            endangered_nonvascular_plants += 1
        elif df['category'].iloc[i] == 'Amphibian':
            endangered_amphibians += 1

        if df['category'].iloc[i] not in endangered_species:
            endangered_species_cnt += 1
            endangered_species.append(df['category'].iloc[i])
    elif df['conservation_status'].iloc[i] == 'Threatened':
        threatened += 1
        if df['category'].iloc[i] == 'Mammal':
            threatened_mammals += 1
        elif df['category'].iloc[i] == 'Bird':
            threatened_birds += 1
        elif df['category'].iloc[i] == 'Reptile':
            threatened_reptile += 1
        elif df['category'].iloc[i] == 'Fish':
            threatened_fishes += 1
        elif df['category'].iloc[i] == 'Vascular Plant':
            threatened_vascular_plants += 1
        elif df['category'].iloc[i] == 'Nonvascular Plant':
            threatened_nonvascular_plants += 1
        elif df['category'].iloc[i] == 'Amphibian':
            threatened_amphibians += 1
    elif df['conservation_status'].iloc[i] == 'In Recovery':
        in_recovery += 1
        if df['category'].iloc[i] == 'Mammal':
            in_recovery_mammals += 1
        elif df['category'].iloc[i] == 'Bird':
            in_recovery_birds += 1
        elif df['category'].iloc[i] == 'Reptile':
            in_recovery_reptile += 1
        elif df['category'].iloc[i] == 'Fish':
            in_recovery_fishes += 1
        elif df['category'].iloc[i] == 'Vascular Plant':
            in_recovery_vascular_plants += 1
        elif df['category'].iloc[i] == 'Nonvascular Plant':
            in_recovery_nonvascular_plants += 1
        elif df['category'].iloc[i] == 'Amphibian':
            in_recovery_amphibians += 1

# print(species_of_concern + endangered + threatened + in_recovery)
# 191

print("Species of Concern: " + str(species_of_concern))
print("Endangered: " + str(endangered))
print("Threatened: " + str(threatened))
print("In Recovery: " + str(in_recovery))

# Output
# Species of Concern: 161
# Endangered: 16
# Threatened: 10
# In Recovery: 4

# print("Endangered species count: " + str(len(endangered_species)))
# 5

# print("Endangered species categories: " + str(endangered_species)) 
# ['Mammal', 'Bird', 'Fish', 'Vascular Plant', 'Amphibian']

# Endangered species distribution
print(str(endangered_mammals) + " Endangered mammals") # 7 Endangered mammals
print(str(endangered_birds) + " Endangered birds") # 4 Endangered birds
print(str(endangered_reptile) + " Endangered reptiles") # 0 Endangered reptiles
print(str(endangered_fishes) + " Endangered fishes") # 3 Endangered fishes
print(str(endangered_vascular_plants) + " Endangered Vascular Plants") # 1 Endangered Vascular Plants
print(str(endangered_nonvascular_plants) + " Endangered NonVascular Plants") # 0 Endangered NonVascular Plants
print(str(endangered_amphibians) + " Endangered amphibians") # 1 Endangered amphibians

# In recovery species distribution
print(str(in_recovery_mammals) + " mammals in recovery") # 1 mammals in recovery
print(str(in_recovery_birds) + " birds in recovery") # 3 birds in recovery
print(str(in_recovery_reptile) + " reptiles in recovery") # 0 reptiles in recovery
print(str(in_recovery_fishes) + " fishes in recovery") # 0 fishes in recovery
print(str(in_recovery_vascular_plants) + " Vascular Plants in recovery") # 0 Vascular Plants in recovery
print(str(in_recovery_nonvascular_plants) + " NonVascular Plants in recovery") # 0 NonVascular Plants in recovery
print(str(in_recovery_amphibians) + " amphibians in recovery") # 0 amphibians in recovery

# Threatened species distribution
print(str(threatened_mammals) + " threatened mammals") # 2 threatened mammals
print(str(threatened_birds) + " threatened birds") # 0 threatened birds
print(str(threatened_reptile) + " threatened reptiles") # 0 threatened reptiles
print(str(threatened_fishes) + " threatened fishes") # 4 threatened fishes
print(str(threatened_vascular_plants) + " threatened Vascular Plants") # 2 threatened Vascular Plants
print(str(threatened_nonvascular_plants) + " threatened NonVascular Plants") # 0 threatened NonVascular Plants
print(str(threatened_amphibians) + " threatened amphibians") # 2 threatened amphibians

# Occurance of varous categories after clean up
mammal_clean = 0
bird_clean = 0
reptile_clean = 0
amphibian_clean = 0
fish_clean = 0
vascular_plant_clean = 0
nonvascular_plant_clean = 0


for i in range(0, len(df['category'])):
    if df['category'].iloc[i] == 'Mammal':
        mammal_clean += 1
    elif df['category'].iloc[i] == 'Bird':
        bird_clean += 1
    elif df['category'].iloc[i] == 'Reptile':
        reptile_clean += 1
    elif df['category'].iloc[i] == 'Amphibian':
        amphibian_clean += 1
    elif df['category'].iloc[i] == 'Fish':
        fish_clean += 1
    elif df['category'].iloc[i] == 'Vascular Plant':
        vascular_plant_clean += 1
    elif df['category'].iloc[i] == 'Nonvascular Plant':
        nonvascular_plant_clean += 1


print("category column length: " + str(mammal_clean + bird_clean + reptile_clean + amphibian_clean + fish_clean + vascular_plant_clean + nonvascular_plant_clean))
print("Summary breakdown")
print("Mammal: " + str(mammal_clean))
print("Bird: " + str(bird_clean))
print("Reptile: " + str(reptile_clean))
print("Amphibian: " + str(amphibian_clean))
print("Fish: " + str(fish_clean))
print("Vascular Plant: " + str(vascular_plant_clean))
print("Nonvascular Plant: " + str(nonvascular_plant_clean))

# Output
# 191
# Mammal: 38
# Bird: 79
# Reptile: 5
# Amphibian: 7
# Fish: 11
# Vascular Plant: 46
# Nonvascular Plant: 5
```
